```julia
Frec1=30 ##Frecuencia
Frec2=20 ##Frecuenca 2
seg=5 ##Tiempo de Muestreo
sps=10 ##Muestras por segundo
t=collect(linspace(0,seg,seg*sps))##Vector de tiempo
y=cos.(2*3.1416*t*Frec1)+cos.(2*3.1416*t*Frec2);
```

```julia
using PyPlot #Usando PyPlot
```

```julia
plot(y) #Plot
```

```julia
frec=abs.(fft(y,1));
```

```julia
frecvec =collect(linspace(0,sps,sps*seg))
length(frecvec)
length(frec)
```

```julia
subplot(2,1,1)
plot(frecvec[1:50],frec[1:50],"-r")
xlabel("Frec [Hz]") 
ylabel("Amplitud") 
subplot(2,1,2)
plot(t,y)
```
